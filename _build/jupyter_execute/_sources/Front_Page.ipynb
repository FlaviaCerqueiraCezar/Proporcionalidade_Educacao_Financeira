# Proporcionalidade e Educação Financeira

Este material propõe atividades voltadas ao ensino da proporcionalidade na Educação Financeira aplicada a situações do cotidiano, com ênfase na análise de orçamentos familiares, a partir do método 50/30/20 de divisão proporcional da renda em categorias (necessidades, desejos e reservas), e no estudo de proporcionalidades a partir de gráficos, porcentagens e relação entre grandezas. As atividades são estruturadas com base na Taxonomia de Bloom, promovendo a progressão cognitiva entre os níveis de compreender, aplicar, analisar e criar, de modo a articular teoria e prática. O conjunto inclui três eixos de aprendizagem:

(1) Planejamento financeiro familiar Estudo e elaboração de orçamentos familiares;

(2) Proporcionalidade na Educação Financeira: aborda a Educação Financeira e grandezas proporcionais diretas e inversas, relacionando-as a funções matemáticas; e

(3) Juros e proporcionalidade.

Dessa forma, pretende-se promover competências financeiras, desenvolvendo o pensamento crítico e autonomia nas tomadas de decisão econômica pelos estudantes, além de contextualizar a aplicação da proporcionalidade em situações do dia a dia.
